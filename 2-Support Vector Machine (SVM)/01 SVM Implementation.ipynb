```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from sklearn.svm import SVC
```

```python
# Getting features and target values
X,y=diabetes.iloc[:,:-1].values,diabetes.iloc[:,-1].values

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
# Scalling all features
scaler=StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
class SVM_classifier:
    def __init__(self,learning_rate=0.01,num_iter=100,lambda_parameter=0.01):
        self.learning_rate = learning_rate
        self.num_iter = num_iter
        self.lambda_parameter = lambda_parameter

    def fit(self,X,y):
        def updateWtsAndBias(X,y):
            m,n=X.shape
            y_mult=y*y_mult

            for idx in range(m):
                condn = y[idx]*(np.dot(self.wts,X[idx])+self.bias)>=1
                if condn:
                    dw=2*self.lambda_parameter*self.wts+self.lambda_parameter
                    db=0
                else:
                    dw=2*self.lambda_parameter*self.wts - y[idx]*X[idx]
                    db=y[idx]
                
                self.wts-=self.learning_rate*dw
                self.bias-=self.learning_rate*db

        m,n=X.shape
        # Initialize the weights randomly with small values
        self.wts=np.random.rand(n)
        self.bias=np.random.rand(1)

        for _ in range(self.num_iter):
            updateWtsAndBias(X,y)

    def predict(self,X):
        y_pred=np.dot(self.wts,X)+self.bias
        return y_pred
```

```python
model=SVC()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
y_pred
```
