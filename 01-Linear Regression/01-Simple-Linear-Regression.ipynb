```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(context='notebook', style='darkgrid',palette='dark', font_scale=1.2)
%matplotlib inline
```

```python
# Set a random seed for reproducibility
np.random.seed(0)

# Generate random data
num_points = 200
X = 2 * np.random.rand(num_points, 1)
y = 4 + 3 * X + np.random.randn(num_points, 1)

# Plot the data
plt.scatter(X, y,color='orange')
plt.xlabel("Independent Variable (X)")
plt.ylabel("Dependent Variable (y)")
plt.title("Randomly Generated Dataset")
plt.show()
```

```python
# Dividing Data points for Testing and Training
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.15,random_state=69)
```

`Aim is to find the equation of best fit line so we can predict the output with minimum error`

```python
class LinearRegression:
    def fit(self, X, y, max_iter=1000, alpha=0.01):
        def gradientDescent(X, y,max_iter,alpha):
            theta = np.random.randn(2)  # random initialization of thetas
            n = len(y)
            for _ in range(max_iter):
                y_pred = theta[0]*X+theta[1]
                temp=(y_pred-y)
                theta[0] -= alpha*((temp*X).sum()/n)
                theta[1] -= alpha*(temp.sum()/n)
            return theta

        self.theta = gradientDescent(X, y, max_iter, alpha)

    def predict(self, X_test):
        try:
            y_pred = self.theta[0]*X_test+self.theta[1]
            return y_pred
        except:
            raise Exception("Fit the Model properly!")
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
my_predictions=model.predict(X_test)

# sk-learn 
from sklearn.linear_model import LinearRegression as LR
sk_model=LR()
sk_model.fit(X_train,y_train)
sk_model_pred=sk_model.predict(X_test)

plt.scatter(X_test, y_test, color="red")
plt.plot(X_test, my_predictions, color="lime", label="my pred")
plt.plot(X_test, sk_model_pred, color="orange", label="sk-learn pred")
plt.legend()
```
